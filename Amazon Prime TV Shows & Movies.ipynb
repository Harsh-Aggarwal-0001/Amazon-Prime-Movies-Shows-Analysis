## 3. ***Data Wrangling***

### Data Wrangling Code

```python
# Write your code to make your dataset analysis ready.
```

```python
# Make copies to avoid modifying original data accidentally
titles_clean = df_titles.copy()
credits_clean = df_credits.copy()

# ==========================================
# 1. HANDLING MISSING VALUES & CLEANING TYPES
# ==========================================

# Text columns: Replace NaN with appropriate placeholders
titles_clean['description'] = titles_clean['description'].fillna('No description available')
titles_clean['age_certification'] = titles_clean['age_certification'].fillna('Unrated')

# Seasons column: It's NaN for Movies. Let's fill with 0 and convert to integer.
titles_clean['seasons'] = titles_clean['seasons'].fillna(0).astype(int)

# Credits: character names can be NaN for crew members (e.g., Directors)
credits_clean['character'] = credits_clean['character'].fillna('N/A (Crew/Self)')


# ==========================================
# 2. HANDLING SCORES & VOTES (IMDb / TMDb)
# ==========================================
# Drop records that have absolutely no rating if your goal is evaluation/recommendation
# Or you can keep them but flag them. Let's filter out titles with 0 IMDb votes to remove noise.
titles_clean['imdb_votes'] = titles_clean['imdb_votes'].fillna(0).astype(int)

# Create a normalized popularity metric if needed, or handle missing scores
# Let's fill missing scores with the median score to avoid losing data while preserving the distribution center
titles_clean['imdb_score'] = titles_clean['imdb_score'].fillna(titles_clean['imdb_score'].median())
titles_clean['tmdb_score'] = titles_clean['tmdb_score'].fillna(titles_clean['tmdb_score'].median())


# ==========================================
# 3. MERGING THE DATAFRAMES (Titles + Credits)
# ==========================================
# We join the datasets using the unique content 'id'
df_merged = pd.merge(
    credits_clean,
    titles_clean,
    on='id',
    how='inner'
)


# ==========================================
# 4. EXPLODING GENRES FOR ADVANCED ANALYSIS
# ==========================================
# If you want to analyze metrics per individual genre (since one movie can have multiple genres):
df_genres_exploded = titles_clean.explode('genres')


# ==========================================
# VERIFICATION
# ==========================================
print("Titles Cleaned Shape:", titles_clean.shape)
print("Merged Dataset Shape:", df_merged.shape)
```
